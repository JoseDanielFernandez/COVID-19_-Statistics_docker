```python
import mysql.connector
import pymysql
import matplotlib.pyplot as plt

connection = pymysql.connect(host='30.10.0.1', port=3306, user='root', password='root', database='covid')
cursor = connection.cursor(pymysql.cursors.DictCursor)

query = "SELECT casos_confirmados, fallecidos, recuperados FROM `casos` WHERE region LIKE 'Capital District'"

cursor.execute(query)
cant=[]
for i in cursor:
    #print(i['region'])
    #lista.append(i['region'])
    confirmados = i['casos_confirmados']
    fallecidos = i['fallecidos']
    recuperados = i['recuperados']
    cant.append(confirmados)
    cant.append(fallecidos)
    cant.append(recuperados)
    
tipo=['Confirmados','Fallecidos','Recuperados']

print('ESTADÍSTICAS DE CONTAGIO DE COVID-19 EN BOGOTÁ:')
print(' ')
print('Casos Confirmados: ', cant[0])
print('Perdonas fallecidas: ', cant[1])
print('Personas recuperadas: ', cant[2])
    
plt.bar(range(3), cant, edgecolor='black')

plt.xticks(range(3), tipo, rotation=60)
plt.title("ESTADÍSTICAS DE CONTAGIO DE COVID-19 EN BOGOTÁ")
plt.ylim(min(cant)-10000, max(cant)+10000)
plt.show()
    
```

```python
import mysql.connector
import pymysql
import matplotlib.pyplot as plt

connection = pymysql.connect(host='30.10.0.1', port=3306, user='root', password='root', database='covid')
cursor = connection.cursor(pymysql.cursors.DictCursor)

query = "SELECT SUM(casos_confirmados), SUM(fallecidos), SUM(recuperados) FROM `casos`"

cursor.execute(query)
cant=[]
for i in cursor:
    confirmados = i['SUM(casos_confirmados)']
    fallecidos = i['SUM(fallecidos)']
    recuperados = i['SUM(recuperados)']
    cant.append(confirmados)
    cant.append(fallecidos)
    cant.append(recuperados)
    
tipo=['Confirmados','Fallecidos','Recuperados']

print('ESTADÍSTICAS DE CONTAGIO DE COVID-19 EN TODA COLOMBIA:')
print(' ')
print('Casos Confirmados: ', cant[0])
print('Perdonas fallecidas: ', cant[1])
print('Personas recuperadas: ', cant[2])
    
plt.bar(range(3), cant, edgecolor='black', color='yellow')

plt.xticks(range(3), tipo, rotation=60)
plt.title("ESTADÍSTICAS DE CONTAGIO DE COVID-19 EN TODA COLOMBIA")
plt.ylim(min(cant)-10000, max(cant)+10000)
plt.show()
```
